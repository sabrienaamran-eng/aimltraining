```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import vectorize
```

```python
df=pd.read_csv('/content/UpdatedResumeDataSet.csv',encoding='utf-8')
print('Shape of Data Frame',df.shape)
df.head(10)
```

```python
print('\n Columns',df.columns.tolist())
```

```python
sample_df=df.sample(10,random_state=42).reset_index(drop=True)
sample_df
```

```python
sample_df=sample_df.rename(columns={'Resume':'Resume_Text'})
sample_df
print(sample_df.shape)
```

```python
vectorizer=TfidfVectorizer(stop_words='english',max_features=5000)
tfidf_matrix=vectorizer.fit_transform(sample_df['Resume_Text'])
tfidf_matrix.shape
```

```python
similartiness_matrix=cosine_similarity(tfidf_matrix)
sim_df=pd.DataFrame(similartiness_matrix,index=sample_df["Category"]+"_"+sample_df.index.astype(str),
                    columns=sample_df["Category"]+"_"+sample_df.index.astype(str))
sim_df.round(2).head(3)
```
