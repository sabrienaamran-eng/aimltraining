NLP PIPELINE

```python
#import Required Libraries
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
our_text="Artificial Intelligence and Machine Laerning are going to change the future of technology!"
```

```python
tokens=word_tokenize(our_text)
#split text into individuals words/tokens
tokens
```

```python
#cleaning
clean_tokens=[re.sub('[^a-zA-Z]','',token.lower())for token in tokens if token.isalpha()]
clean_tokens
```

```python
#stop Word Removal
stop_words=set(stopwords.words('english'))
filtered_tokens=[token for token in clean_tokens if token not in stop_words]
#Removes common less meaningful words
filtered_tokens
```

```python
#Stemming
stemmer=PorterStemmer()
stemmed_tokens=[stemmer.stem(token) for token in filtered_tokens]
#Reduce the words to their root form (not always a word from dictionary)
stemmed_tokens
```

```python
lemmatizer=WordNetLemmatizer()
lemmatized_tokens=[lemmatizer.lemmatize(token) for token in filtered_tokens]
#Convert words to their dictionary form(lemma) using POS: part of speech
lemmatized_tokens
```

Bag of Words (BOW)

```python
from nltk import corpus
corpus=[
    "Artificial Intelligence is transforming Industries",
    "Machine Learning is a brand of artificial intelligence",
    "Deep Learning is a subfield of machine learning",
    "Natural Language Processing is a subfield of artificial intelligence",
    ]
```

```python
cv=CountVectorizer()
bow_matrix=cv.fit_transform(corpus)
print("**********Features Names (BOW)**********")
print(cv.get_feature_names_out())
print("**********BOW Matrix Array**********")
print(bow_matrix.toarray())
#
```
