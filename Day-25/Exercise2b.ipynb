**Practice Exercise One**

input_text="Natural Language Processing is going to help you out in near future. You have 2 remember this!! for getting interview calls"

Do the following Tasks:

Splits text into words/tokens (show the individual word token)
Clean tokens (show clean token)
Remove Stop words (show token after removing stop words)
Stemming (show the words in their roots forms)
lemmatization (show lemma words)

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
import re

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
input_text="Natural Language Processing is going to help you out in near future. You have 2 remember this!! for getting interview calls"
```

```python
tokens=word_tokenize(input_text)
tokens
```

```python
#Clean Tokens (show clean token)
clean_tokens=[re.sub(r'[^a-zA-Z]','',token.lower()) for token in tokens if token.isalpha]
clean_tokens
```

```python
#removes stop words
stop_words=set(stopwords.words('english'))

filtered_tokens=[token for token in clean_tokens if token not in stop_words]
print('token after removing stop words')
filtered_tokens
```

```python
stemmer=PorterStemmer()
stemmed_tokens=[stemmer.stem(token) for token in filtered_tokens]
print('The Tokens in their root forms')
stemmed_tokens
```

```python
#lemmatization
lemmatizer=WordNetLemmatizer()
lemmatized_tokens=[lemmatizer.lemmatize(token) for token in filtered_tokens]
print('Lemma Words')
lemmatized_tokens
```

```python
tfidf=TfidfVectorizer()
matrix=tfidf.fit_transform (lemmatized_tokens)
matrix
df=pd.DataFrame(matrix.toarray(),columns=tfidf.get_feature_names_out())
df
```

```python
print('\n Top Weighted Terms')
print(df.mean().sort_values(ascending=False).head(10) )
```

**TF-IDF Example-2**

```python
corpus=["Python developer required with strong coding skills",
        "Experience in REST API, Automation, Jenkins, Flask and Python",
        "Looking for machine learning engineer familiar with NLP, Supervised, Unsupervised and Deep Learning"
        ]
```

```python
tfdif=TfidfVectorizer()

matrix=tfdif.fit_transform(corpus)
df=pd.DataFrame(matrix.toarray(),columns=tfdif.get_feature_names_out(10))
df
```

```python
print('\n Top Weight Terms')
print(df.mean().sort_values(ascending=False).head(10))
```
