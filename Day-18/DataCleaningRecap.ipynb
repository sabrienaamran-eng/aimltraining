```python
import pandas as pd
```

```python
# sample dataset
data = {
    'Name': ['Ali','Sara','John',None,'Marry'],
    'Age': [25, None, 30, 22, None],
    'City':['KL', 'Penang', None, 'Johor', 'KL']
}
```

```python
df=pd.DataFrame(data)
print('Original DataFrame:')
print(df)
```

# Check the missing values Row by Row

```python
for index, row in df.iterrows():
    print(f"Row {index}: Missing = {row.isnull().any()}, Details={row.isnull().to_dict()}")
```

```python
# show only rows with missing data
missing_rows=df[df.isnull().any(axis=1)]
print('Rows with missing data:')
print(missing_rows)
```

```python
# Simulate rows that would be dropped uising dopna()
to_drop=df[df.isnull().any(axis=1)]
print(to_drop)

df_cleaned=df.dropna()
print('\nAfter dropna():')
print(df_cleaned)
```

```python
df_filled=df.fillna({
    'Name': 'Unknown',
    'Age':df['Age'].mean(),
    'City': 'Not Available'
})
```

```python
df_cleaned=df.dropna()
print('\nAfter dropna():')
print(df_cleaned)
```
