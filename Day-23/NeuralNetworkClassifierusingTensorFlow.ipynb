**Predict Heart Disease**


1.   Load DataSet
2.   Split features and Test
3.   Train Trest Spil 4. Scale Features



```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
heart_data=pd.read_csv('/content/sample_data/heart.csv')
heart_data.head()
```

```python
x=heart_data.drop('target',axis=1) #drop means except
y=heart_data['target']
print(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# 1. Creating the Model
model=tf.keras.Sequential(
    [
        tf.keras.layers.Dense(16, activation='relu',input_shape=(x_train.shape[1],)),
        tf.keras.layers.Dense(8, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ]
)
```

```python
# 2. Compiling the Model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 3. Train the Model
history_model=model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
#
```

```python
# 4. Evaluate the Model
loss,accuracy=model.evaluate(x_test,y_test)
print(f'Loss:{loss},Accuracy:{accuracy}')
```

**Example 2**

*Bank Customer Exit Prediction.

*   Based on given data using Neural Network Binary Classifier find out is a customer going to stay or leave stay: (Churn=0) and leave (churn=1)
*   Predict the result for new data as well at end

```python
#Example
bank_data=pd.read_csv('/content/sample_data/bank_churn.csv')
bank_data.head()
```

```python
le=LabelEncoder()
bank_data['Gender']=le.fit_transform(bank_data['Gender'])
bank_data=pd.get_dummies(bank_data,columns=['Geography'])
x=bank_data.drop(['Exited'],axis=1)
y=bank_data['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
bank_model=tf.keras.Sequential(
    [
        tf.keras.layers.Dense(16, activation='relu',input_shape=(x_train.shape[1],)),
        tf.keras.layers.Dense(8, activation='relu'),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ]
)
```

```python
bank_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
bank_model_history=bank_model.fit(x_train,y_train,epochs=100,validation_data=(x_test,y_test))
```

```python
loss,accuracy=bank_model.evaluate(x_test,y_test)
print(f'Loss:{loss},Accuracy:{accuracy}')
```

```python
# Predict a new customer
new_customer=pd.DataFrame({
        'CreditScore':[720],
        'Geography':['France'],
        'Gender':['Male'],
        'Age':[42],
        'Balance':[62000],
        'NumOfProducts':[1],
        'EstimatedSalary':[55000]
    })

#Encode and scale same as before
new_customer['Gender']=le.transform(new_customer['Gender'])
new_customer=pd.get_dummies(new_customer,columns=['Geography'],drop_first=True)

#Add missing dummy columns
for col in [c for c in x.columns if c not in new_customer.columns]:
  new_customer[col]=0

new_customer=new_customer[x.columns] #same order
new_scaled=scaler.transform(new_customer)

#prediction
pred=bank_model.predict(new_scaled)
if pred[0][0]>0.5:
  print("Customer will Leave")
  print(pred[0][0])
else:
  print("Customer will Stay")
  print(pred[0][0])
```
