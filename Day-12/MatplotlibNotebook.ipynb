    ## Figure: The complete window to create plot
    ## Axes: The area where data is plotted
    ## Axis: x line and y line to define scale and ticks

# First Example to create a line chart from our data
    ## Students.xlsx for fetching the data
    ## Generate line chart based on Student Name and Student Marks

```python
import pandas as pd
```

```python
df=pd.read_excel('C:/Users/sabri/OneDrive/Desktop/aimltraining/Day-10/students.xlsx')
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='s',color='green')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(names,marks,marker='o',color='orange')
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
print('Solid Line')
plt.figure(figsize=(6,4))
plt.plot(names,marks,marker='o',color='orange',linestyle='-')
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
print('Dashed Line')
plt.plot(names,marks,marker='o',color='orange',linestyle='-')
plt.show()
```

```python
print('Dotted Line')
plt.plot(names,marks,marker='o',color='orange',linestyle=':')
plt.show()
```

```python
print('Dashed Dotted Line')
plt.plot(names,marks,marker='o',color='orange',linestyle='-.')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(names,marks,marker='v',color='blue',linestyle='-.')
plt.title('Student Progress Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```
