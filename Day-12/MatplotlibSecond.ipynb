# Scatter Plots
## In scatter plots werepresent data in points, where each point represent a data observation, We can check how to values correlate

```python

product_sales={'Laptop':750,'Smart Phone':650,'Smart Watch':400,'Tablets':500,'AC':400}
products=list(product_sales.keys())
sales=list(product_sales.values())
plt.scatter(products,sales)
plt.title('Basic Product and Sales Representation')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid('True')
plt.show()
```

```python
players=['MSD','Rohit','Virat','Gill','Shami']
match1_score=[45,67,56,78,89]
match2_score=[55,78,40,80,97]
match3_score=[90,40,20,30,60]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='red',label='Match One', s=100)
plt.scatter(players,match2_score,color='green',label='Match Two', s=100, marker="d")
plt.scatter(players,match3_score,color='blue',label='Match Three', s=150, marker="p")
plt.title('Comparision of Players in Three Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.legend(loc='lower right')

plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```
