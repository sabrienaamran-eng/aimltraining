```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
x=np.array([
    [1, 3500],
    [2, 3200],
    [3, 3000],
    [4, 2500],
    [5, 2300],
    [6, 2000],
    [7, 2200],
    [2, 3700],
])
y= ['Unfit', 'Unfit', 'Unfit', 'Fit', 'Fit', 'Fit', 'Fit', 'Unfit']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(x,y)
```

```python
new_person=[[4.5, 2600]] # 4.5 hrs exercise, 2600 calories
pred=model.predict(new_person)
print("Predicted Result:",pred[0])
```

```python
exercise_hours=float(input('Enter Number of Exercise Hours'))
cal_intake=float(input('Enter number of calories consumed'))
pred_fu=model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours {exercise_hours}, Calories Intake {cal_intake} Your are likely {pred_fu[0]}')
```

```python
import pandas as pd
```

```python
df=pd.read_csv('fruit_data.csv')
df
```

```python
x=df[['Weight','Size']]
y=df['Fruit']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(x,y)
```

```python
new_fruit=[[120, 2]] # 190 gm size is 9 cm
pred=model.predict(new_fruit)
print("Predicted Result:",pred[0])
```

```python
fruit_weight=float(input('Enter weight of fruit in gm'))
fruit_size=float(input('Enter size of fruit in cm'))
our_fruit=pd.DataFrame([[fruit_weight, fruit_size]], columns=['Weight','Size'])
pred_fruit=model.predict(our_fruit)
print(f'Fruit Weight is {fruit_weight}, fruit_size is {fruit_size} The Fruit is likely an: {pred_fruit[0]}')
```
