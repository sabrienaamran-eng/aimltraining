# Supervised Machine Learning
## Linear Regresssion

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data=pd.read_csv('sales_data.csv')
data.head()
```

```python
x=data[["Advertising_Spend"]]
y=data["Sales_Revenue"]
```

```python
model=LinearRegression()
model.fit(x,y)
print('Training Done') #tak letak keluar kotak
```

```python
new_spend=float(input('Enter New Spend to Predict Sales Revenue'))
predicted_revenue=model.predict([[new_spend]])
print(f'New Spend {new_spend} Predicted Revenue {predicted_revenue}')
```

## 1. Linear Regression - House Price Prediction
## Download house_price.csv file from github
** Goal:** Train a Linear Regression model to predict house price from Size_sqft

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('house_price.csv')
df.head(5)
```

```python
s_sqft=df[['Size_sqft']]
p_sqft=df['Price']
print(s_sqft)
print(p_sqft)
```

```python
model=LinearRegression()
```

```python
model.fit(s_sqft,p_sqft)
```

```python
price_per_sqft=model.predict([[1700]])
print(price_per_sqft)
```
