# Matplotlib.pyplot.subplots()
### Subplots(): provides a way to plot multiple plots on single figure.
### Given number of rows and columns. It returns a tuple (fig,ax) giving a single figure fig with an array of axes as.

```python
import matplotlib.pyplot as plt
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
```

```python
import numpy as np
```

```python
sales_data={'Electronics':45000,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
actual_sale=list(sales_data.values())
print(category)
print(actual_sale)
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.random(10)
ax1.plot(x,x/2)
ax2.scatter(np.random.random(20),np.random.random(20))
ax3.bar(category,actual_sale)
ax4.hist(np.random.randn(100))
```

# Subplots using indexing

```python
our_lines=np.linspace(0,10,100)
```

```python
product_prices={'Hair oil':20,'Face Wash':30, 'Lip Balm':10, 'Face Cream':15}
```

```python
our_numbers=np.random.random(10)
```

```python
our_dots=np.random.randn(100)
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['blue','green','red','orange'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
ax[1,1].hist(our_dots,color='purple')
plt.tight_layout()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,5))
ax[0,0].plot(our_lines,our_lines/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color=['blue','green','red','orange'])
ax[1,0].scatter(np.random.randn(100),np.random.randn(100),color='orange')
ax[1,1].hist(our_dots,color='purple')

# first row 0 & first column 1 means [0,0] Line Plot
#---------------------------------------------------------------------------------------
ax[0,0].set_title('Line Plot Example')
ax[0,0].set_xlabel('X Axix')
ax[0,0].set_ylabel('Y Axis')
ax[0,0].grid(True,linestyle='--',alpha=0.7,color='lightblue',axis='y')
# first row 0 & second column 1 means [0,1] Line Plot
#---------------------------------------------------------------------------------------
ax[0,1].set_title('Product Sales Information')
ax[0,1].set_xlabel('Product Name')
ax[0,1].set_ylabel('Product Sale')
ax[0,1].grid(True,linestyle='--',alpha=0.7,color='lightgreen',axis='y')
#---------------------------------------------------------------------------------------

# second row 1 & first column 0 means [1,0] Line Plot
#---------------------------------------------------------------------------------------
ax[1,0].set_title('Our Scatter Plot')
ax[1,0].set_xlabel('Product Name')
ax[1,0].set_ylabel('Product Sale')
ax[1,0].grid(True,linestyle='--',alpha=0.7,color='pink',axis='y')
# second row 1 & first column 1 means [1,1] Line Plot
#---------------------------------------------------------------------------------------
ax[1,1].set_title('Histogram Example')
ax[1,1].set_xlabel('Value Range')
ax[1,1].set_ylabel('Frequency')
ax[1,1].grid(True,linestyle='--',alpha=0.7,color='orange',axis='y')


fig.suptitle('Matplot Lib SubPlots example using 2 rows and 2 cols')
plt.tight_layout()
plt.savefig('Subplotex.jpeg',dpi=300,bbox_inches='tight')
```
