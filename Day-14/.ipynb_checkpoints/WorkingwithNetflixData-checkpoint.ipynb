# Use Netflix dataset to visualize the distribution of release years for Movies Vs TV Shows using Seaborn's histplot()
## Use different color for Movie Vs TV Show

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('netflix_data.csv')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(
    data=df,x='release_year',hue='type',multiple='stack',     
    palette={'Movie': 'red', 'TV Show': 'blue'},
    bins=40
)
plt.title("Distribution of Release Years for Movies vs TV Shows")
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
