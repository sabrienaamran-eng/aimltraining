Exercise Create two tensors with random values between 0 and 1 , shape is [6,300]
**Find out the following details: **
Number of elements
Size,
Rank,
Shape


```python
import tensorflow as tf

# Create two tensors with random values between 0 and 1, shape [6, 300]
tensor_a = tf.random.uniform(shape=[6, 300], minval=0, maxval=1)
tensor_b = tf.random.uniform(shape=[6, 300], minval=0, maxval=1)

# Display the tensors
print("Tensor A:\n", tensor_a)
print("\nTensor B:\n", tensor_b)

# Find details for Tensor A
print("\n--- Details for Tensor A ---")
print("Shape:", tensor_a.shape)
print("Number of elements:", tf.size(tensor_a).numpy())
print("Rank:", tf.rank(tensor_a).numpy())
print("Size in bytes:", tensor_a.numpy().nbytes)

# Find details for Tensor B
print("\n--- Details for Tensor B ---")
print("Shape:", tensor_b.shape)
print("Number of elements:", tf.size(tensor_b).numpy())
print("Rank:", tf.rank(tensor_b).numpy())
print("Size in bytes:", tensor_b.numpy().nbytes)
```
